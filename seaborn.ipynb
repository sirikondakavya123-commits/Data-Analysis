```python
## 4. Histograms & Countplots
# Histogram
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

```python
### 5. Scatter Plots
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

```python
### 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
```

```python
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

```python
### 7. Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
### 8. Heatmap
# Correlation heatmap (numeric columns only)
#thick red color ->strongly correlated
#neg->neg correlated
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
 
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='violet')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
```

```python
#1. Compare Passenger Survival by Class and Gender
import seaborn as sns
df = sns.load_dataset("titanic")
sns.countplot(data=df, x="class", hue="survived")
plt.title("Passenger Distribution by Class ")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()


sns.countplot(data=df, x="sex", hue="survived")
plt.title("Passenger Distribution by Gender")
plt.xlabel("Passenger sex")
plt.ylabel("Count")
plt.show()
```

```python
#2. Distribution of Passenger Ages
df = sns.load_dataset("titanic")
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
#3. Fare Distribution by Class
df = sns.load_dataset("titanic")
sns.histplot(data=df, x='fare', bins=20, kde=True, color='skyblue')
plt.title("Fare Distribution by Class")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
#4. Relationship Between Age and Fare
df = sns.load_dataset("titanic")
sns.scatterplot(data=df, x='age', y='fare', hue='class', palette='viridis')
plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
```

```python
#5. Correlation Heatmap
df = sns.load_dataset("titanic")
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```
