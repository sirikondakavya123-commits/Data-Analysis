```python
import matplotlib.pyplot as plt
```

```python
#line plot
x = [1,2,3,4,5]
y = [10,20,25,30,40]
 
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
#bar plot
categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.bar(categories, values, color='purple')
plt.title("Bar Plot Example")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
#scatter plot
x = [5,7,8,7,2,17,2,9,4,11]
y = [99,86,87,88,100,86,103,87,94,78]
 
plt.scatter(x, y, color='red', marker='.')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
data = [34, 65, 8, 99, 67]
plt.hist(data, bins=5, color="blue", edgecolor="black")
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
#histogram
data = [22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]
 
plt.hist(data, bins=5, color='orange', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
#customized plot
x = [1,2,3,4,5]
y = [2,3,5,7,11]
 
plt.plot(x, y, color='purple', linestyle='--', marker='*', label='Line 1')
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5,9,9]
y = [2, 3, 5, 7, 11,13,10]

plt.subplot(1, 2, 1)
plt.plot(x, y, label="Prime Numbers", color='blue', marker='o')
plt.title("subplot Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#pie chart
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [215, 130, 245, 210]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
 
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
 
```

```python
#used to find outliers visually
data = [25, 30, 35, 35, 40, 45, 50, 55, 60]
plt.boxplot(data)
plt.title("Box Plot Example (Statistical Distribution)")
plt.ylabel("Values")
plt.show()
```

```python
#violin plot
import numpy as np
data = np.random.normal(0, 1, 100)
plt.violinplot(data)
plt.title("Violin Plot Example")
plt.show()
 
 
```

```python
#stem plot
x = np.linspace(0, 2*np.pi, 10)
y = np.cos(x)
plt.stem(x, y)
plt.title("Stem Plot Example")
plt.show()
```

```python
#step plot
x = [1,2,3,4,5]
y = [2,3,5,6,8]
plt.step(x, y, where='mid', color='magenta')
plt.title("Step Plot Example")
plt.show()
 
```

```python
#Area plot
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.fill_between(x, y, color='lightblue', alpha=0.6)
plt.title("Area Plot Example")
plt.show()
```

```python
#contour plot
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
 
plt.contour(X, Y, Z, cmap='plasma')
plt.title("Contour Plot Example")
plt.colorbar()
plt.show()
```

```python
#filled contour plot
plt.contourf(X, Y, Z, cmap='grey')
plt.title("Filled Contour Plot")
plt.colorbar()
plt.show()
```

```python
#polar plot
theta = np.linspace(0, 77*np.pi, 100)
r = np.abs(np.sin(2*theta))
plt.polar(theta, r)
plt.title("Polar Plot Example")
plt.show()
 
```

```python
#error bar plot
x = np.arange(1,6)
y = np.random.randint(5,15,5)
errors = np.random.rand(5)
 
plt.errorbar(x, y, yerr=errors, fmt='o', ecolor='red', capsize=5)
plt.title("Error Bar Plot Example")
plt.show()
```

```python
#horizontal bar plot
categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Plot")
plt.show()
 
```

```python
#stack plot

days = [1,2,3,4,5]
eating = [2,3,4,3,4]
sleeping = [8,7,8,6,7]
working = [8,9,7,8,8]
 
plt.stackplot(days, eating, sleeping, working, labels=['Eating','Sleeping','Working'])
plt.legend()
plt.title("Stack Plot Example")
plt.show()
```

```python
#hexbin plot
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize=25, cmap='plasma')
plt.colorbar()
plt.title("Hexbin Plot Example")
plt.show()
```

```python
#3d surface plot
from mpl_toolkits.mplot3d import Axes3D
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.plot_surface(X, Y, Z, cmap='hsv')
ax.set_title("3D Surface Plot")
plt.show()
```

```python
#3d line plot
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
t = np.linspace(0, 100, 1000)
x = np.sin(t)
y = np.cos(t)
z = t
 
ax.plot3D(x, y, z, 'purple')
ax.set_title("3D Line Plot")
plt.show()
```

```python
#3d scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
 
ax.scatter(x, y, z, color='red', marker='o')
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
#3d wireframe plot
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_title("3D Wireframe Plot")
plt.show()
```

```python
#3d counter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour3D(X, Y, Z, 50, cmap='plasma')
ax.set_title("3D Contour Plot")
plt.show()
```

```python
#3d bar plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = np.zeros(5)
dx = np.ones(5)
dy = np.ones(5)
dz = [5, 2, 7, 3, 6]
 
ax.bar3d(x, y, z, dx, dy, dz, color='teal')
ax.set_title("3D Bar Plot")
plt.show()
```

```python
#3D Surface with Colormap
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))
 
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')
fig.colorbar(surf)
ax.set_title("3D Surface with Colormap")
plt.show()
 
```

```python
#3D Sphere Surface
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='skyblue')
ax.set_title("3D Sphere Surface")
plt.show()
```

```python
#3D Surface + Wireframe
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7, cmap='viridis')
ax.plot_wireframe(X, Y, Z, color='black', linewidth=0.5)
ax.set_title("3D Surface + Wireframe")
plt.show()
```

```python
#3D Contour Projection on Planes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour(X, Y, Z, zdir='z', offset=-2, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='x', offset=-5, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='y', offset=5, cmap='coolwarm')
 
ax.set_title("3D Contour Projection on Planes")
plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [98, 78, 87, 94, 89]
students = ['a', 'b', 'c', 'd', 'e']
plt.plot(students, marks, marker='o', color='blue')
plt.title("Line Plot of Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#2.Plot a bar chart of sales of 4 products.
#bar plot
products = ['pen','pencil','chocolate','cookies']
sales = [150,120,200,140]
 
plt.bar(products,sales, color='purple')
plt.title("sales of 4 product")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
#3.Make a scatter plot of Age vs Salary.
#scatter plot
age=[24,45,35,36,21,19,27,34]
salary=[35000,90000,80000,82000,25000,0,45550,60000]
 
plt.scatter(age, salary, color='red', marker='.')
plt.title("scatter plot of Age vs Salary.")
plt.xlabel("age")
plt.ylabel("salary")
plt.grid(True)
plt.show()
```

```python
#4.Draw a histogram of marks scored by students in a class.
marks=[98,78,99,100,56,32,45,27]
plt.hist(marks,bins=8,color="lavender",edgecolor="black")
plt.title("histogram of marks scored by students in a class.")
plt.xlabel("marks")
plt.ylabel("Frequency")
plt.show()
```

```python
#5.Customize a plot with different colors, markers, and line styles.
x=[34,56,88,9,33,90,33]
y=[34,88,5,3,77,3,45]
 
plt.plot(x, y, color='violet', linestyle='--', marker='.', label='Line')
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
#6.Create two subplots: Line plot and bar plot side by side.
x=[34,56,88,9,33,90,33]
y=[34,88,5,3,77,3,45]
 
plt.subplot(1, 2, 1)
plt.plot(x, y, label="plot", color='blue', marker='o')
plt.title("line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(1,2,2)
plt.bar(x,y,color="cyan")
plt.title("bar plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
